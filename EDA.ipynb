```python
stat_df = pd.read_csv('data/PLANTS/ABORIGINAL-PICULET.csv')
stat_df.head()
```

```python
stat_df['TIME'] = pd.to_datetime(stat_df['DATE_TIME']).dt.time
stat_df.head()
```

```python
from datetime import datetime
operation_hours = 0

for i in range(stat_df.shape[0]):
    if stat_df['CMP_SPEED_2'][i] == 0 and stat_df['CMP_SPEED_2'][i+1] > 0:
        start_time = stat_df['TIME'][i+1]
    if stat_df['CMP_SPEED_2'][i] > 0 and stat_df['CMP_SPEED_2'][i+1] == 0:
        end_time = stat_df['TIME'][i]
    
    print(type(start_time), end_time)
    if(start_time and end_time):
        operation_hours += datetime.strptime(start_time, '%H:%M:%S') - datetime.strptime(end_time, '%H:%M:%S')

operation_hours
```
